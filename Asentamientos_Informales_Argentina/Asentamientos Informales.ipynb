Luego de conocer como esta compuesto el DF. se comienza a trabajar en el mismo para llegar a tener una base mas util para analizar.

```python
df.drop(['otros_nombres_barrio', 'IDENTIFICAR', 'cosas_que_mas_le_gustan1'], axis = 'columns', inplace=True)
# Eliminamos columnas que no voy a utilizar en el analisis porque no son relevantes o no tienen datos significativos para analizar.

# inplace=True la eliminación de valores se hace directamente en el dataframe original.
```

```python
# Hay ciertos lugares o valores en el dataframe que no contiene valor o son NAN, entonces para no tener que anular una fila por no tener
# disponible esos valores, va a ser segun que datos (NAN) tenga la columna, los transformo en un dato mas significativo para el analisis

df["tipo_de_barrio"].fillna("VILLA", inplace = True) # Con esto relleno el NAN con el valor VILLA. El inplace=True modifica el dataset original
df["nombre_barrio"].fillna("Otro", inplace = True)
df["pendiente"].fillna("Plano", inplace = True)
df["anio_creo_barrio"].fillna("No sabe", inplace = True)
df["localidad"].fillna("Caba", inplace = True)
df["inundaciones"].fillna("No tuvo", inplace = True)
df["basural"].fillna("No tiene", inplace = True)
df["actividad_agropecuaria"].fillna("No tiene", inplace = True)
df["emergencias_ambulancia"].fillna("No tiene", inplace = True)
df["agua"].fillna("No tiene", inplace = True)
df["relleno_sanitario"].fillna("No tiene", inplace = True)
df["camino_de_alto_trafico"].fillna("No tiene", inplace = True)
df["desechos_industriales"].fillna("No sabe", inplace = True)
df["plantacion_forestal"].fillna("No sabe", inplace = True)
df["sala_medica"].fillna("No tiene", inplace = True)
df["emergencias_bomberos"].fillna("No tiene", inplace = True)
df["ribera_rio"].fillna("No tiene", inplace = True)
df["comisaria"].fillna("No tiene", inplace = True)
df["torres_de_alta_tension"].fillna("No tiene", inplace = True)
df["linea_de_tren"].fillna("No tiene", inplace = True)
df["organizaciones_externas"].fillna("No tiene", inplace = True)
```

```python
# Ahora a centrar el estudio en los valores de luz, asfalto y alumbrado.

data1 = pd.get_dummies(data, columns=['asfalto'])
```

```python
import urllib.request
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

import numpy as np
```

```python
# Cargamos los datos y los dividimos en conjuntos para entrenar y testear.
data = data1
y = data["asfalto_SÍ"]
X = data.drop(["asfalto_SÍ"], axis=1)
(X_train, X_test,
 y_train, y_test) = train_test_split(X,
                                     y,
                                     test_size=0.33,
                                     random_state=42)
```

```python
# Primero creamos un arbol de decisión para ver la precisión y poder compararla
tree = DecisionTreeClassifier(max_depth=2, random_state=11) # Max depth acorta el arbol si no salia muy extenso.
tree.fit(X_train, y_train)
print(f"La precisión del Arbol de decisión es de: {tree.score(X_test, y_test)}")
```
