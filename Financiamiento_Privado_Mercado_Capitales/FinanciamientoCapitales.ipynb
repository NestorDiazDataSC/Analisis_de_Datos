Financiamiento Privado en Mercado de Capitales por tamaño de empresa e instrumento, monto en dólares.

```python
# Comenzamos importando las librerias a utilizar por el momento.

import numpy as np   # Importamos librerias de calculo
import pandas as pd  # Importamos librerias de uso de tablas o los dataframes
import seaborn as sns # Importamos librerias de graficos
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Ubicar el dataset, cargarlo y mostrarlo.
url='https://raw.githubusercontent.com/NestorDiazDataSC/Analisis_de_Datos/main/Financiamiento_Privado_Mercado_Capitales/financiamiento-privado-mercado-capitales-dolares.csv'

df= pd.read_csv(url)
df
```

```python
df.columns # Para ver las columnas del Dataset, para esta parte del analisis voy a usar las primeras 13 columnas que tienen datos sobre el financiamiento.
```

```python
# Elimino las columnas que no voy a utilizar del dataset
df = df.drop(df.columns[13:77], axis=1)
```

```python
df.columns # Verifico que quedaron solo las columnas que necesito
```

```python
df.columns = ['Fecha', 'ON GE', 'FF GE', 'ACC GE', 'CHPD GE', 'TOTAL GE', 'ON PyMES', 'FF PyMES', 'ACC PyMES', 'CHPD PyMES','PAG PyMES','TOTAL PyMES', 'TOTAL']
```

Descripcion de las columnas del dataset, los valores numericos estan expresados en dolares.

*   ON = Obligaciones negociables emitidas, monto
*   FF = Fideicomisos Financieros, emitidos
*   ACC = Acciones emitidas
*   CHPD = Cheques de pago diferidos emitidos
*   PAG = Pagares emitidos
*   GE = Grandes Empresas
*   PyMES = Pequeñas y medianas empresas

```python
df.isnull().sum() # Para ver los valores nulos de cada columna.
```

```python
# En estos casos de abajo, reemplazo los valores nulos por "0" puesto que son nulos porque no hubo emisión en esos casos.
df["ON GE"].fillna(0, inplace = True)
df["FF GE"].fillna(0, inplace = True)
df["ACC GE"].fillna(0, inplace = True)
df["CHPD GE"].fillna(0, inplace = True)
df["ON PyMES"].fillna(0, inplace = True)
df["FF PyMES"].fillna(0, inplace = True)
df["ACC PyMES"].fillna(0, inplace = True)
df["CHPD PyMES"].fillna(0, inplace = True)
df["PAG PyMES"].fillna(0, inplace = True)
# En estos casos no puedo poner "0" porque son totales, sumas de varios valores, por lo que relleno con esas sumas.
df["TOTAL GE"].fillna(df["ON GE"] + df["FF GE"] + df["ACC GE"] + df["CHPD GE"], inplace = True)
df["TOTAL PyMES"].fillna(df["ON PyMES"] + df["FF PyMES"] + df["ACC PyMES"] + df["CHPD PyMES"] + df["PAG PyMES"], inplace = True)
df["TOTAL"].fillna(df["TOTAL GE"] + df["TOTAL PyMES"], inplace = True)
```

```python
df.isnull().sum() # Para ver los valores nulos de cada columna.
```

```python
# Ahora vemos que nos trae el booleano FALSE porque ya no hay mas valores NAN en el dataframe y podemos seguir trabajando.
check_for_nan = df.isnull().values.any()
print (check_for_nan)
```

#NORMALIZACION DE DATOS

```python
# Primer codigo para reducir el valor, para expresarlo en millones.
# Segundo renglon de codigo, para convertir numeros que estan en "notación cientifica" a numeros para que sea mas facil de interpretar y tambien que sean "Float"
df ["ON GE"] = df ["ON GE"] / 1000
df['ON GE'] = df['ON GE'].apply(lambda x: float('{:.0f}'.format(x)))

df ["FF GE"] = df ["FF GE"] / 1000
df['FF GE'] = df['FF GE'].apply(lambda x: float('{:.0f}'.format(x)))

df ["ACC GE"] = df ["ACC GE"] / 1000
df['ACC GE'] = df['ACC GE'].apply(lambda x: float('{:.0f}'.format(x)))

df ["CHPD GE"] = df ["CHPD GE"] / 1000
df['CHPD GE'] = df['CHPD GE'].apply(lambda x: float('{:.0f}'.format(x)))

df ["TOTAL GE"] = df ["TOTAL GE"] / 1000
df['TOTAL GE'] = df['TOTAL GE'].apply(lambda x: float('{:.0f}'.format(x)))

df ["ON PyMES"] = df ["ON PyMES"] / 1000
df['ON PyMES'] = df['ON PyMES'].apply(lambda x: float('{:.0f}'.format(x)))

df ["FF PyMES"] = df ["FF PyMES"] / 1000
df['FF PyMES'] = df['FF PyMES'].apply(lambda x: float('{:.0f}'.format(x)))

df ["ACC PyMES"] = df ["ACC PyMES"] / 1000
df['ACC PyMES'] = df['ACC PyMES'].apply(lambda x: float('{:.0f}'.format(x)))

df ["CHPD PyMES"] = df ["CHPD PyMES"] / 1000
df['CHPD PyMES'] = df['CHPD PyMES'].apply(lambda x: float('{:.0f}'.format(x)))

df ["PAG PyMES"] = df ["PAG PyMES"] / 1000
df['PAG PyMES'] = df['PAG PyMES'].apply(lambda x: float('{:.0f}'.format(x)))

df ["TOTAL PyMES"] = df ["TOTAL PyMES"] / 1000
df['TOTAL PyMES'] = df['TOTAL PyMES'].apply(lambda x: float('{:.0f}'.format(x)))

df ["TOTAL"] = df ["TOTAL"] / 1000
df['TOTAL'] = df['TOTAL'].apply(lambda x: float('{:.0f}'.format(x)))
```
